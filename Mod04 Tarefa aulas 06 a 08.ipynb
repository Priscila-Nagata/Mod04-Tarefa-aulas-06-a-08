####  2) Melhorando a interação com o API
Observe dois URLs de consultas diferentes, por exemplo o URL utilizado em aula, e o URL feito no exercício anterior. Compare-os e observe as diferenças.

1. Faça uma função em Python que recebe como argumento o UF da consulta e o tipo de interessado (```'Estado'```ou ```Município```), e que devolve os dados da consulta no formato *DataFrame*.
2. Quantas solicitações para o Estado podem ser consultadas para Minas Gerais com *status* em 'Arquivado por decurso de prazo' estão registradas?
3. Qual é o município da Bahia com mais solicitações deferidas?
4. Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

````python
#1) Seu código aqui

def formatar_tipo(tipo):

    tipo_formatado = tipo.capitalize()
    if tipo_formatado.upper() == 'MUNICIPIO':
        tipo_formatado = 'Município'
    return tipo_formatado


def filtrar_dados(df_items, sigla_uf, tipo):
    sigla_uf = sigla_uf.upper()
    tipo = formatar_tipo(tipo)
    
    if tipo.lower() in ["estado", "estados"]:
        tipo = "Estado"
    
    return df_items.loc[(df_items['uf'] == sigla_uf) & (df_items['tipo_interessado'] == tipo)]

sigla_uf = input("Olá! Digite a uf desejada (somente a sigla): ")
tipo = input("Você deseja consultar informações do Estado ou de um Município: ")


df_items_resposta = filtrar_dados(df_items, sigla_uf, tipo)


df_items_resposta
````

````python
# 2) Seu código aqui

df_items.loc[(df_items['uf'] == 'MG') & (df_items['tipo_interessado'] == 'Município') & (df_items['status'] == 'Arquivado por decurso de prazo')].value_counts('status')
````

````python
# 3) Seu código aqui

tipos_deferido = ['Deferido','Deferido (PVL-IF)','Deferido (decisão judicial)']

df_ba_deferidos = df_items.loc[
                      (df_items['uf'] == 'BA') & 
                      (df_items['tipo_interessado'] == 'Município') & 
                      (df_items['status'].isin(tipos_deferido))]

df_ba_deferidos['interessado'].value_counts().sort_values(ascending=False)


````

````python
# 4) Seu código aqui
#Salve um arquivo .csv com os dados de solicitações da Bahia, com interessado = 'Estado'

df_ba_estado = filtrar_dados(df_items, 'BA', 'Estado')

df_ba_estado.head(3)
````

````python
df_ba_estado.to_csv('df_ba_estado.csv', encoding='utf-8',index= False, sep=';', decimal=',')
````
